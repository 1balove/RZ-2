**Спецификая 1 с core CPI**

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv('clear_data/overall_data.csv')
```

```python
data
```

```python
data['g_t_lag1'] = data['core_cpi'].shift(1)
data['g_t_lag2'] = data['core_cpi'].shift(2)
data['g_t_lag3'] = data['core_cpi'].shift(3)
data['g_t_lag4'] = data['core_cpi'].shift(4)
data['g_t_lead4'] = data['core_cpi'].shift(-4)
```

```python
data['diff_t'] = data['core_cpi'] - data['g_t_lag1']
data['diff_t1'] = data['g_t_lag1'] - data['g_t_lag2']
data['diff_t2'] = data['g_t_lag2'] - data['g_t_lag3']
data['diff_t3'] = data['g_t_lag3'] - data['g_t_lag4']
data['y_t'] = data['g_t_lead4'] - data['core_cpi']
```

```python
data = data.dropna()
data
```

```python
X = data[['diff_t', 'diff_t1', 'diff_t2', 'diff_t3']]
X = sm.add_constant(X)  # свободный член
y = data['y_t']
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())
```

```python
regression = model.predict(X)
regression 
```

```python
data['spec1_task4'] = regression
data[['year', 'quarter', 'spec1_task4']].to_csv('clear_data/spec1_task4.csv')
```
