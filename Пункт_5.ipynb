```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('clear_data/overall_data.csv')
data
```

```python
data['inf_t+4'] = data['headline_cpi'].shift(-4)
data
```

```python
data1 = data.dropna()
data1
```

```python
X = sm.add_constant(data1['inf_expectation'])
y = data1['inf_t+4']
model_spec3 = sm.OLS(y, X).fit()
print(model_spec3.summary())
```

```python
spec3_pred = model_spec3.predict(X)
spec3_pred
```

```python
data1['spec3_pred'] = spec3_pred
data1
```

```python
data1[['year', 'quarter', 'spec3_pred']].to_csv('clear_data/spec3_res.csv')
```
