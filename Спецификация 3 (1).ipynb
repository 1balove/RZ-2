```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv('overall_data.csv')
```

```python
data
```

```python
p = data['inf_expectation']
p
```

```python
data['inf_t+4'] = data['headline_cpi'].shift(-4)
```

```python
data
```

```python
data1=data.dropna(subset=['inf_t+4'])
```

```python
data1
```

```python
X = sm.add_constant(data1['inf_expectation'])
y = data1['inf_t+4']
model_spec3 = sm.OLS(y, X).fit()
print(model_spec3.summary())
```

```python
data3=pd.DataFrame(model_spec3.predict(X))
data3
```
